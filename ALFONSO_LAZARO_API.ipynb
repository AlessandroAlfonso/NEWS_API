# Getting data from an API
This notebook walks you through some steps in collecting data from The Verge News using the News API.


### Import package

```python
from newsapi import NewsApiClient
from datetime import datetime
from datetime import timezone

import pandas as pd
```

### Get articles via News API

```python
pd.set_option('display.max_colwidth', None)

news_list = NewsApiClient(api_key = 'f563f261c4f741a28f330d7db0a312b5')

articles = news_list.get_everything(sources = 'the-verge',
                                     from_param = '2022-06-08',
                                     to = '2022-06-09',
                                     language = 'en',
                                     sort_by = 'popularity',
                                     )

news = articles.get('articles')
```

```python
newsapi_list = pd.DataFrame(news)
```

```python
newsapi_list.columns = ['source', 'author', 'title', 'description', 'url', 'urlToImage', 'Published','content' ]
```
